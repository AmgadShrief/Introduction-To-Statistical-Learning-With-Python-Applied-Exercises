# First We Will See some Numpy Functions that is really helpful

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Read And Clean The Data

```python
college = pd.read_csv('/College.csv')
college
```

```python
college = college.rename({"Unnamed: 0" : "College Name"}, axis = 1)
college
```

```python
college = college.set_index("College Name")
```

```python
college
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']], diagonal = None);
```

```python
college.boxplot(column = 'Outstate',by = 'Private')
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,50,100], labels=['No', 'Yes'])
college
```

```python
college['Elite'].value_counts()
```

```python
college.boxplot(column = 'Outstate',by = 'Elite')
```

```python
fig,ax = plt.subplots(2,2,figsize=(8,8),sharey=True)
ax[0][0].hist(college['Outstate'],bins = 20)
ax[0][1].hist(college['Room.Board'],bins = 20)
ax[1][0].hist(college['Books'],bins = 20)
ax[1][1].hist(college['Personal'],bins = 20)
ax[0][0].set_title('Outstate')
ax[0][1].set_title('Room & Board')
ax[1][0].set_title('Books')
ax[1][1].set_title('Personal');
```

```python
fig2,ax2 = plt.subplots(1,2,figsize=(12,4))
ax2[0].scatter(college['Outstate'],college['Expend'])
ax2[1].scatter(college['Outstate'],college['Grad.Rate'])
ax2[0].set_xlabel('Outstate')
ax2[0].set_ylabel('Expend')
ax2[1].set_xlabel('Outstate')
ax2[1].set_ylabel('Grad Rate');
```

from this plot we can observe a positive correlation between higher out-of-state tuition fees ('Outstate') and both the instructional expenditure per student ('Expend') and the graduation rate ('Grad Rate').
